```python
blocks = gpd.read_file("il-data/IL_NBM_Blocks_SHP _Jun 2014/IL_Block.shp").to_crs(epsg = 3528)
wireless = gpd.read_file("il-data/IL_NBM_Wireless_SHP_June 2014/IL_Wireless.shp").to_crs(epsg = 3528)
```

```python
verizon = wireless.query("DBANAME == 'Verizon Wireless'")
att     = wireless.query("DBANAME == 'AT&T Mobility LLC'")
tmobile = wireless.query("DBANAME == 'T-Mobile'")
sprint  = wireless.query("DBANAME == 'Sprint'")

verizon
```

```python
fig, ax = plt.subplots(1, 4, figsize = (7.5, 3), subplot_kw = {"aspect" : "equal"})

att    .plot(alpha = 0.15, color = "blue",    ax = ax[0])
verizon.plot(alpha = 0.2,  color = "red",     ax = ax[1])
tmobile.plot(alpha = 0.15, color = "magenta", ax = ax[2])
sprint .plot(alpha = 0.5,  color = "gold",    ax = ax[3])

for axi in ax:
    map_format(axi)
    axi.set_xlim(ax[0].get_xlim())
    axi.set_ylim(ax[0].get_ylim())

fig.savefig("wireless_2014.png", dpi = 300)
```
